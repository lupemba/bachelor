```python
%matplotlib notebook
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
omni = pd.read_pickle('/home/simon/Desktop/Bachelor_project/data/pkl_files/omni.pkl')
```

```python
omni.head()
```

```python
active = np.ones(len(omni.Kp))
active[omni.Kp<4] = float('NaN')
max_dens = max(omni.Kp)
min_dens = min(omni.Kp)
plt.figure()

p1 = omni.Kp.plot( style='b',legend=True,linewidth=0.7)
p1.fill_between(omni.Kp.index,max_dens* active, np.zeros(len(active))+min_dens, color='r', alpha=0.4)
p1.set_ylabel('Kp')
plt.title('Geomagnetic active periods')
```
