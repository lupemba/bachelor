```python
%matplotlib notebook
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from apexpy import Apex
```

```python
start = Apex(date=FAC.index[0].date())  # datetime objects are also supported
finish = Apex(date=2030.0)
```

```python
h=FAC.Radius/1000-6371
```

```python
h_1= h
lat = FAC.Latitude.values
long = FAC.Longitude.values
```

```python
FAC.index.values
```

```python
start = Apex(date=FAC.index[0].date()) 
mlat1, mlon1 = start.geo2apex(lat, long, h_1)
mlt1= start.mlon2mlt(mlon1, FAC.index.values, ssheight=450)
```

```python
finish = Apex(date=FAC.index[-1].date())
mlat2, mlon2 = finish.geo2apex(lat, long, h_1)

mlt2= start.mlon2mlt(mlon2, FAC.index.values, ssheight=450)
```

```python
plt.figure()

plt.plot(mlat1-mlat2)
plt.title('Effect of the choice of time mlat')
```
