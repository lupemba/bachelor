```python
%matplotlib inline
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
DNS = pd.read_pickle('/home/simon/Desktop/Bachelor_project/data/pkl_files/DNS.pkl')
omni = pd.read_pickle('/home/simon/Desktop/Bachelor_project/data/pkl_files/omni.pkl')
```

```python
DNS.head()
```

```python
omni.head()
```

```python
omni = omni.reindex(DNS.index,method='ffill')
DNS = pd.concat([DNS,omni], axis=1)
```

```python
DNS.head()
```

```python
DNS_quiet =  DNS[DNS.Kp < 4]
```

```python
y_quiet = DNS_quiet.Density
y = DNS.Density

x1_quiet = (DNS_quiet.Altitude.values)[:, np.newaxis]
x1 = (DNS.Altitude.values)[:, np.newaxis]

x2_quiet = (DNS_quiet.F10_Mean.values+DNS_quiet.F10_INDEX.values)[:, np.newaxis]/2
x2 = (DNS.F10_Mean.values+DNS.F10_INDEX.values)[:, np.newaxis]/2
```

```python
model_alt = LinearRegression(fit_intercept=True)
model_alt.fit(x1_quiet, y_quiet)

print(model_alt.intercept_)
print(model_alt.coef_)

dens_alt_quiet = y_quiet-model_alt.predict(x1_quiet)
```

```python
plt.figure()

plt.scatter(x1_quiet, 10**12*y_quiet, marker='o')
plt.plot(x1_quiet,10**12*model_alt.predict(x1_quiet),'r')
plt.title('Dens vs altitude regression')
```

```python
model_f10 = LinearRegression(fit_intercept=True)
model_f10.fit(x2_quiet,dens_alt_quiet)

print(model_f10.intercept_)
print(model_f10.coef_)
```

```python
plt.figure()

plt.scatter(x2_quiet, 10**12*dens_alt_quiet, marker='o')
plt.plot(x2_quiet,10**12*model_f10.predict(x2_quiet),'r')
plt.title('Dens_alt vs F10 regression')
```

```python
dens_alt_f10 = y-model_alt.predict(x1)-model_f10.predict(x2)
```

```python
plt.figure()

plt.scatter(DNS.Kp.values, 10**12*dens_alt_f10, marker='o')
```

```python
plt.figure()

p1 = Half_orbit[Half_orbit.Hemisphere==1].Density_res.plot(legend=True)
p2 = Half_orbit[Half_orbit.Hemisphere==1].Density.plot(secondary_y=True,legend=True,style='k')
#p2 = Half_orbit[Half_orbit.Hemisphere==1].AP_INDEX.plot(secondary_y=True, style='r',legend=True)
p1.set_ylabel('kg/m^3')
p1.right_ax.set_ylabel('F10.7')
plt.title('Half orbital means')
```
