```python
my.add_apex_coords(DNS,date = 'none', h = 450)
my.add_heading(DNS, latitude = 'mLatitude' )
```

```python
mOrbit_nr = np.cumsum(np.append(0, abs(np.diff(DNS.N_heading.values))))/4
```

```python
plt.figure()

plt.plot(np.array(range(len(DNS.Orbit_nr.values))),DNS.Orbit_nr.values)
plt.plot(np.array(range(len(DNS.Orbit_nr.values))),mOrbit_nr)
```

```python
start_orbit=450

N = 100

fig_index = np.arange(90,-90,-0.5)

fig_orbit = np.array(range(N))/2+start_orbit

fig_values = np.zeros([len(fig_index),len(fig_orbit)])

for i in range(N):
    data_column = DNS[abs(mOrbit_nr-start_orbit+i/2)<0.01]
    data_column = data_column.set_index('mLatitude')
    fig_values[:,i] = data_column.reindex(labels=fig_index, method='nearest', tolerance=1).Density.values
```

```python
plt.figure()

plt.imshow(fig_values, extent=[0, 5, 0, 5], origin='lower',
           cmap='jet')
plt.colorbar()
plt.axis(aspect='image')
```
